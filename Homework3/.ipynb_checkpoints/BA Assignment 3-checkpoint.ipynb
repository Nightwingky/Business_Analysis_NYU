### 1.4 Using the CustomerID column, identify the number of customer who have done recurring purchases. 

```r
library(sqldf)
df<-data.cheesemakers
recurring<-sqldf("select [Customer.ID], count([Customer.ID]) as count from df 
    group by [Customer.ID] having count([Customer.ID])>1")
head(recurring)
```

* What is the average number of purchases of the recurring clients?

```r
mean(recurring$count)
```

* What is the average spent by recurring clients?

```r
average_spent<-sqldf("select avg([Sale.amount]) from df where [Customer.ID] in (
    select [Customer.ID] from df group by [Customer.ID] having count([Customer.ID])>1)")
average_spent
```

* What is the variance in gross profits between recurring clients vs clients who buy 1 cheese

```r
var_recurring<-sqldf("select variance([Gross.profit]) as var_recurring from df where [Customer.ID] in (
    select [Customer.ID] from df group by [Customer.ID] having count([Customer.ID])>1)")
var_one<-sqldf("select variance([Gross.profit]) as var_one from df where [Customer.ID] in (
    select [Customer.ID] from df group by [Customer.ID] having count([Customer.ID])=1)")
```

```r
var_recurring
```

```r
var_one
```

### 1.5 Which are the most profitable clients?

```r
profit<-sqldf("select [Customer.ID], sum([Gross.profit]) as total_profit from df 
    group by [Customer.ID] order by total_profit desc")
head(profit)
```

### 1.6 How many clients are paying more than 2 standard deviations of the mean price? What does that mean in english?

```r
unusual<-sqldf("select [Customer.ID] from df where [Sale.amount] > (
    select (avg([Sale.amount]) + 2 * stdev([Sale.amount])) from df
    ) group by [Customer.ID]")
length(unusual$Customer.ID)
```

### 1.7 Compute number of unique clients per state

```r
state_record<-sqldf("select [State], count(distinct([Customer.ID])) as customer_num, 
    sum([Number.of.Records]) as records, sum([Sale.amount]) as total_sales from df group by [State];")
head(state_record)
```

* Normalize the data using min-max scaling

```r
mean_num<-mean(state_record$customer_num)
mean_records<-mean(state_record$records)
mean_sales<-mean(state_record$total_sales)
state_record$customer_num<-(state_record$customer_num - min(state_record$customer_num)) / 
    (max(state_record$customer_num) - min(state_record$customer_num))
state_record$records<-(state_record$records - min(state_record$records)) / 
    (max(state_record$records) - min(state_record$records))
state_record$total_sales<-(state_record$total_sales - min(state_record$total_sales)) / 
    (max(state_record$total_sales) - min(state_record$total_sales))
head(state_record)
```

* Is there an association (correlation) between client volume and sales?

```r
cor(state_record$customer_num, state_record$records)
cor(state_record$customer_num, state_record$total_sales)
```

### 2.2 Compared to the 95% confidence interval, is the 90% confidence interval for toy sales wider or narrower? Please explain your answer very briefly. 

```r
names(data.store)[4]<-"toy"
toy_sd<-sd(data.store$toy)
toy_mean<-mean(data.store$toy)
SE<-toy_sd/sqrt(length(data.store$toy))
```

```r
confidence_90max<-toy_mean+qnorm(0.95)*SE
confidence_90min<-toy_mean-qnorm(0.95)*SE
confidence_90max
confidence_90min
```

```r
con95max<-toy_mean+qnorm(.975)*SE
con95min<-toy_mean-qnorm(.975)*SE
con95max
con95min
```
